```python
import matplotlib.pyplot as plt
import random

# plt.hist([1,1,2,3,4,5,5,5,5,7])
plt.style.use('default')
# 주사위 던지기
dice = []
for i in range(10):
    dice.append(random.randint(1, 6))

print(dice)

plt.hist(dice, bins=6)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
# with open("datas/temp_seoul.csv", "r", encoding="utf-8") as f:
with open("datas/temp_seoul.csv", "r") as f:
    data = csv.reader(f)  # 데이터 읽어온 자료 저장
    next(data)
    temp_high = []
    for row in data:
        if row[-1] != '':
            row[0] = row[0].replace('\t', '')  #년도앞 탭(공백) 제거
            if row[0].split('-')[1] == "08": #연도를 정수화 비교
                temp_high.append(float(row[-1]))
                
                
print(len(temp_high))
# print(temp_high)
plt.rc('font', family='Malgun Gothic')
plt.title("서울의 8월 최고기온 분포")
plt.hist(temp_high, color='r', bins=100)
plt.xlabel('온도')
plt.ylabel('빈도수')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# plt.style.use('_mpl-gallery')

# make data
np.random.seed(1)
x = 4 + np.random.normal(0, 1.5, 200)

# plot:
# fig, ax = plt.subplots()
fig, ax = plt.subplots(figsize=(8, 6))  # fig 크기 설정

ax.hist(x, bins=8, linewidth=0.5, edgecolor="white")

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 56), yticks=np.linspace(0, 56, 9))

# 추가적인 fig 설정
fig.suptitle('Histogram of Random Data', fontsize=16)  # 제목 추가
fig.tight_layout(pad=3.0)  # 레이아웃 조정

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Fixing random state for reproducibility
np.random.seed(19680801)

plt.style.use('bmh')


def plot_beta_hist(ax, a, b):
    ax.hist(np.random.beta(a, b, size=10000),
            histtype="stepfilled", bins=25, alpha=0.8, density=True)


fig, ax = plt.subplots()
plot_beta_hist(ax, 10, 10)
plot_beta_hist(ax, 4, 12)
plot_beta_hist(ax, 50, 12)
plot_beta_hist(ax, 6, 55)
ax.set_title("'bmh' style sheet")

plt.show()
```
