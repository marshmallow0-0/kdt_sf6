# 내 생일의 최고 기온 분석 및 시각화

```python
# 최고 기온만 리스트에 담기
import csv
import matplotlib.pyplot as plt
# with open("datas/temp_seoul.csv", "r", encoding="utf-8") as f:
with open("datas/temp_seoul.csv", "r") as f:
    data = csv.reader(f)
    next(data)

    temp_high = [] #최고 기온
    temp_low = [] #최저 기온
    for row in data:
        if row[-1] != '':
            row[0] = row[0].replace('\t', '')
            if int(row[0].split('-')[0]) >= 1980: #연도를 정수화 비교
                
                if row[0].split('-')[1] == "05" and row[0].split('-')[2] == "27":
                    temp_high.append(float(row[-1]))
                    temp_low.append(float(row[-2]))

# print(len(temp_high))
# print(temp_high)

plt.rc('font', family="Malgun Gothic")
plt.title("내 생일의 최고 기온과 최저 기온")
plt.xlabel("연도별 기온")
plt.ylabel("온도")

plt.plot(temp_high, color="r", label = 'high')
plt.plot(temp_low, color="b",  label = 'low')
plt.legend()
plt.style.use()

plt.rcParams['axes.unicode_minus'] = False

plt.show()
    
```
