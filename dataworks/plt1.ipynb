#### 시각화 라이브러리 사용 - matplotlib

```python
import matplotlib.pyplot as plt

plt.title("Basic Graph")
# plt.plot([1,2,3,4]) # x축값이 생략되면 y축값만 사용된다.
plt.plot([1,2,3,4], [10,40,20,30], color='k') #가로(x) 축 값, 세로(y)축 값
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rc('font', family="Malgun Gothic")
plt.title("시험 성적 그래프", pad=10, loc="right") #pad(패딩) - 여백, 10 -> 10px
plt.plot([1,2,3,4], [90,80,85,70]) #가로(x) 축 값, 세로(y)축 값
plt.xlabel('학생번호') #x축 제목
plt.ylabel('시험점수') #y축 제목

# 격자 형태 
axis = {'both', 'x', 'y'}

# plt.grid() #default - 가로, 세로 표시
plt.grid(True, axis='y')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rc('font', family="Malgun Gothic")
plt.title("범례 및 색상 표시")
plt.plot([10,20,30,40], label="비만율", color='r') #가로(x) 축 값, 세로(y)축 값
plt.plot([40,30,20,10], label="출생율", color='g') #가로(x) 축 값, 세로(y)축 값
plt.legend() #범례(소제목) 필수 명시
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rc('font', family="Malgun Gothic")
plt.title("범례 및 색상 표시")
plt.plot([10,20,30,40], 'b.', label="circle") #가로(x) 축 값, 세로(y)축 값
plt.plot([40,30,20,10],'r^', label="triangle") #가로(x) 축 값, 세로(y)축 값
plt.legend() #범례(소제목) 필수 명시
plt.show()
```

```python
import matplotlib.pyplot as plt

x = list(range(6))
y = [1,2,3,5,6,7]

plt.bar(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = range(3)
y = [100, 400, 500]
colors = ['C1', 'C2', 'C3']
# 세로 막대 그래프
plt.bar(x, y, color=colors, width = 0.6) # width- 막대 너비

# 눈금
plt.xticks(x, ['2018', '2019', '2020'])
plt.show(), 
```

```python
import matplotlib.pyplot as plt

y = range(3)
x = [100, 400, 500]
colors = ['r', 'g', 'b']
# 가로 막대 그래프
plt.barh(y, x, color=colors, height = 0.6) # width- 막대 너비

# 눈금
plt.yticks(y, ['2018', '2019', '2020'])
plt.show()
```
