```python
import cv2 
from pytesseract import pytesseract
import matplotlib.pyplot as plt 

# 이미지를 흑백으로 불러오기 

img = cv2.imread("source/car1.png", cv2.IMREAD_GRAYSCALE)
# img = cv2.imread("source/sonata1.png", cv2.IMREAD_GRAYSCALE)

# 이미지에서 밝기가 35 이하인 모든 픽셀은 검정(0)이 되고, 그보다 큰 픽셀들은 흰색(255)
#이 경우, 픽셀 값이 임계값(35)보다 클 때 그 픽셀 값은 255(흰색)이 됩니다.
_, bin_img = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY)
# _, bin_img = cv2.threshold(img, 35, 255, cv2.THRESH_BINARY)

plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()

pytesseract.tesseract_cmd = \
    r"C:\Program Files\Tesseract-OCR\tesseract.exe"

# 이미지 -> 텍스트로 변환
text = pytesseract.image_to_string(bin_img, lang="kor")
print(text)
```
