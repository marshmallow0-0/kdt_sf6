### OCR(광학문자인식) - 이미지에 있는 텍스트를 추출해서 인식함
### 필수 라이브러리 - 테세랙트(tesseract), 파이테세랙트(pytesseract)

- Tesseract는 OCR 엔진입니다. 이미지에서 텍스트를 인식하고 추출하는 작업을 수행
- C:\Program Files\Tesseract-OCR
- Pytesseract는 Tesseract의 Python 래퍼입니다. 즉, Tesseract 엔진을 Python 환경에서 쉽게 사용할 수 있도록 만든 라이브러리

```python
import cv2 
from pytesseract import pytesseract
import matplotlib.pyplot as plt 

# 이미지를 흑백으로 불러오기 

img = cv2.imread("source/news.png", cv2.IMREAD_GRAYSCALE)

plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
```

```python
pytesseract.tesseract_cmd = \
    r"C:\Program Files\Tesseract-OCR\tesseract.exe"

# 이미지 -> 텍스트로 변환
text = pytesseract.image_to_string(img, lang="kor+eng")
print(text)
```
